```python
import pandas as pd
import numpy as np
import pickle
from scipy.io import loadmat
from numpy.linalg import norm
```

```python
# 16-faces to produce similarity matrix
mat = loadmat('simu8_data/simu8_Experiment1.mat')
mdata = mat['events']
face_corrd = mdata['facecoordinates']
face_corrd_list = []
for x in face_corrd:
    if len(x[0].tolist()) != 0:
        face_corrd_list.append(x[0].tolist()[0])
face_corrd_16 = np.unique(face_corrd_list,axis=0)
face_corrd_16 = [np.array(x) for x in face_corrd_16]
face_corrd_16
```

```python
face_sim = np.zeros((16,16))
for i in range(16):
    for j in range(16):
        A = face_corrd_16[i]
        B = face_corrd_16[j]
        cosine = np.dot(A,B)/(norm(A)*norm(B))
        face_sim[i,j] = (cosine + 1) / 2
face_sim
```

```python
s_mat = np.zeros((32,32))
np.fill_diagonal(s_mat,1)
s_mat[0:16,0:16] = face_sim
s_mat
```

```python
np.save("simu8_data/simu8_smat.npy",s_mat)
```

### Distance

```python
face_distance = np.zeros((16,16))
for i in range(16):
    for j in range(16):
        A = face_corrd_16[i]
        B = face_corrd_16[j]
        distance = norm(A-B)
        face_distance[i,j] = distance.round(4)
face_distance
```

```python
np.save("siu8_data/simu8_distance.npy",face_distance)
```
