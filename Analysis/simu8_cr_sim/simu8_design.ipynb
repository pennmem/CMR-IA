```python
# simplified! one list per session.
import pandas as pd
import numpy as np
import pickle
```

```python
face_list = ['Face' + str(x) for x in range(1,17)]
name_list = ['Jim','John','Rob','Bill','Dave','Rich','Charles','Joe','Tom','Chris','Dan','Paul','Mark','Mike','George','Ken']
item_list = face_list + name_list
item_list = np.array(item_list)
```

```python
simu_sess_num = 10000
list_num = 1
n = 8
facepool = np.arange(1,17)
namepool = np.arange(17,33)
serial_pos = np.arange(8)
```

```python
df_study = pd.DataFrame()
df_test = pd.DataFrame()
rng = np.random.default_rng(seed=42)

for sess in range(simu_sess_num):
    sess_faces = rng.choice(facepool, n, replace = False)
    sess_names = rng.choice(namepool, n, replace = False)
    sess_pairs = np.vstack([sess_faces,sess_names]).T

    for lst in range(list_num):
        tmp_df_study = pd.DataFrame()
        tmp_df_test = pd.DataFrame()

        # choose present words
        pres_words = rng.permutation(sess_pairs)
        tmp_df_study['study_itemno1'] = pres_words[:,0]
        tmp_df_study['study_itemno2'] = pres_words[:,1]
        tmp_df_study['study_item1'] = item_list[pres_words[:,0]-1]
        tmp_df_study['study_item2'] = item_list[pres_words[:,1]-1]
        tmp_df_study['serial_pos'] = serial_pos
        tmp_df_study['list'] = lst
        tmp_df_study['session'] = sess
        df_study = pd.concat([df_study,tmp_df_study])
        
        # choose test words
        study_pos = rng.permutation(serial_pos)
        test_probe = []
        correct_ans = []
        for pos in study_pos:
            test_probe.append(pres_words[pos][0])
            correct_ans.append(pres_words[pos][1])
        tmp_df_test['test_itemno'] = test_probe
        tmp_df_test['test_item'] = item_list[np.array(test_probe)-1]
        tmp_df_test['correct_ans'] = correct_ans
        tmp_df_test['study_pos'] = study_pos
        tmp_df_test['list'] = lst
        tmp_df_test['session'] = sess
        df_test = pd.concat([df_test,tmp_df_test])

df_study.reset_index(inplace = True, drop=True)
df_test.reset_index(inplace = True, drop=True)
```
