```python
import scipy.io
import numpy as np
import pandas as pd
import pickle

np.set_printoptions(threshold=np.inf)
```

```python
# select 984 words from 1638
mat = scipy.io.loadmat("../wordpools/concreteness_imageability_norms.mat", squeeze_me=True)
c = mat["C"]
# print(c)
selected_idx = np.where(c > 0)[0]
```

```python
# 1638 words
with open("../wordpools/wasnorm_wordpool.txt") as f:
    items = f.readlines()
items = [i.replace("\n", "") for i in items]
```

```python
# the 984 words
selected_item = [items[i] for i in selected_idx]
```

```python
# the 984 itemno
selected_itemno = [i + 1 for i in selected_idx]
```

```python
# frequency
mat = scipy.io.loadmat("../wordpools/Frequency_norms.mat", squeeze_me=True)
frequency = mat["F"]
selected_freq = [frequency[i] for i in selected_idx]
```

```python
# the 984 word dataframe
data = {"item": selected_item, "itemno_old": selected_itemno, "itemno": np.arange(1, 985), "freq": selected_freq}
df = pd.DataFrame(data=data)
df
```

```python
# define bins as Lonhas and Kahana (2013)
bins = [1, 36, 68, 115, 163, 235, 344, 495, 816, 1575, 26215]
df["quantile"] = pd.cut(df.freq, bins, labels=np.arange(10)).astype(int)
df
```

```python
df.groupby("quantile").freq.mean().round()
```

```python
# save
with open("simu4_data/simu4_word_freq.pkl", "wb") as outp:
    pickle.dump(df, outp, pickle.HIGHEST_PROTOCOL)
```
