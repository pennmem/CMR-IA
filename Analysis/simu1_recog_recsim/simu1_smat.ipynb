```python
import pandas as pd
import numpy as np
import pickle
```

```python
from gensim import models
```

```python
import gensim.downloader as api
```

```python
with open("simu1_data/simu1_design.pkl", 'rb') as inp:
    df = pickle.load(inp)
df
```

```python
items = np.unique(df.item.to_numpy())
item2no = {}
for i in range(len(items)):
    item2no[items[i]] = i+1
wordlist = list(item2no.keys())
wordlist = [i.lower() for i in wordlist]
wordlist
```

```python
wv = api.load('fasttext-wiki-news-subwords-300')
```

```python
# lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
```

```python
s_mat = np.zeros((len(wordlist),len(wordlist)))
for i in range(len(wordlist)):
    for j in range(len(wordlist)):
        s_mat[i,j] = wv.similarity(wordlist[i],wordlist[j])
s_mat
```

```python
np.save('simu1_data/simu1_300_smat.npy',s_mat)
```
